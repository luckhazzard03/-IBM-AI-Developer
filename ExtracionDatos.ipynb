```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Paso 2: Obtener el nombre del archivo subido
file_name = list(uploaded.keys())[0]  # Tomar el primer archivo subido
```

```python
# Paso 3: Verificar nombres de las hojas disponibles
xls = pd.ExcelFile(file_name)  # Cargar archivo Excel
print("Hojas disponibles en el archivo:", xls.sheet_names)
```

```python
# Paso 4: Cargar el archivo Excel con el nombre correcto de la hoja
df = pd.read_excel(file_name, sheet_name="Año 1992")  # Cambiar si el nombre es diferente
```

```python
# Paso 5: Filtrar registros donde la columna B es 1 (presente) y 0 es(no presente)
df_presentes = df[df.iloc[:, 1] == 1]
df_no_presentes = df[df.iloc[:,1]== 0]
```

```python
# Paso 6: Contar cuántos registros hay con 1 y cuantos 0 no estan presentes
cantidad_presentes = len(df_presentes)
cantidad_no_presentes = len(df_no_presentes)
print(f"Total de registros presentes en 1992: {cantidad_presentes}")
print(f"Total de registros No presentes en 1992: {cantidad_no_presentes}")
```

```python
# Paso 7: Mostrar los consecutivos de la columna A (consecutivos de los registros presentes y no presentes)
consecutivos_presentes = df_presentes.iloc[:, 0]  # Acceder a la columna A para los presentes
consecutivos_no_presentes = df_no_presentes.iloc[:, 0]  # Acceder a la columna A para los no presentes
# Mostrar los consecutivos
print("\nConsecutivos de los registros presentes en 1992:")
print(consecutivos_presentes)

print("\nConsecutivos de los registros NO presentes en 1992:")
print(consecutivos_no_presentes)
```

```python
# Gráfico de barras
labels = ['Presentes', 'No Presentes']  # Etiquetas para el gráfico
values = [cantidad_presentes, cantidad_no_presentes]  # Valores de los registros

# Crear gráfico de barras
plt.bar(labels, values, color=['green', 'red'])

# Añadir título y etiquetas
plt.title('Número de Registros Presentes vs No Presentes en 1992')
plt.xlabel('Estado')
plt.ylabel('Cantidad de Registros')

# Mostrar gráfico
plt.show()

# Si deseas graficar los consecutivos también
# Graficar los consecutivos de la columna A de los registros presentes
plt.figure(figsize=(10, 6))
plt.plot(consecutivos_presentes, label='Consecutivos Presentes', marker='o', linestyle='-', color='green')
plt.plot(consecutivos_no_presentes, label='Consecutivos No Presentes', marker='x', linestyle='-', color='red')

# Añadir título y etiquetas
plt.title('Consecutivos de Registros Presentes vs No Presentes')
plt.xlabel('Índice')
plt.ylabel('Consecutivo')

# Añadir leyenda
plt.legend()

# Mostrar gráfico
plt.show()
```
