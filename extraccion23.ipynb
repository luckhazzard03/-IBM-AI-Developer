```python
import pandas as pd
from google.colab import files
```

```python
# 📌 Subir los archivos Excel
print("📂 Sube el archivo con las hojas de ANALES...")
uploaded_1 = files.upload()

print("📂 Sube el archivo con la hoja LIBROS ENTREGADOS...")
uploaded_2 = files.upload()
```

```python
# 📌 Obtener los nombres de los archivos subidos
file_1 = list(uploaded_1.keys())[0]  # Archivo con ANALES
file_2 = list(uploaded_2.keys())[0]  # Archivo con LIBROS ENTREGADOS
```

```python
# 📌 Cargar los archivos Excel
xls_1 = pd.ExcelFile(file_1)
xls_2 = pd.ExcelFile(file_2)
```

```python
# 📌 Definir las hojas a extraer de cada archivo
hojas_anales = ['ANALES CAMARA', 'ANALES SENADO', 'ANALES CONGRESO']
hoja_libros = 'LIBROS ENTREGADOS'

# 📌 Definir las columnas de interés en mayúsculas
columnas_anales = ["SUBSERIE", "RANGO", "FECHA INICIAL", "FECHA FINAL"]
columnas_libros = ["NOMBRE", "RANGO", "FECHA INICIAL", "FECHA FINAL"]

# 📌 Diccionario para almacenar los datos extraídos
datos_extraidos = {}

# 📌 Extraer datos de ANALES
for hoja in hojas_anales:
    if hoja in xls_1.sheet_names:
        print(f"\n📊 Procesando hoja: {hoja}...")
        df = pd.read_excel(xls_1, sheet_name=hoja)

        # ✅ Normalizar los nombres de columnas
        df.columns = df.columns.str.strip().str.upper()

        # ✅ Filtrar solo las columnas necesarias
        columnas_existentes = [col for col in columnas_anales if col in df.columns]
        df_seleccionado = df[columnas_existentes]

        # ✅ Reemplazar valores vacíos por 0, excepto en fechas
        df_seleccionado = df_seleccionado.applymap(lambda x: 0 if pd.isna(x) and not isinstance(x, pd.Timestamp) else x)

        # ✅ Almacenar los datos
        datos_extraidos[hoja] = df_seleccionado

        print(df_seleccionado.head())
    else:
        print(f"\n⚠️ La hoja '{hoja}' no está en el archivo. Se omite.")

# 📌 Extraer datos de LIBROS ENTREGADOS y filtrar por NOMBRE
if hoja_libros in xls_2.sheet_names:
    print(f"\n📚 Procesando hoja: {hoja_libros}...")
    df_libros = pd.read_excel(xls_2, sheet_name=hoja_libros)

    # ✅ Normalizar los nombres de columnas
    df_libros.columns = df_libros.columns.str.strip().str.upper()

    # ✅ Filtrar solo las filas donde "NOMBRE" sea ANALES SENADO, CAMARA o CONGRESO
    df_filtrado = df_libros[df_libros["NOMBRE"].isin(hojas_anales)]

    # ✅ Filtrar solo las columnas necesarias
    columnas_existentes_libros = [col for col in columnas_libros if col in df_filtrado.columns]
    df_libros_seleccionado = df_filtrado[columnas_existentes_libros]

    # ✅ Reemplazar valores vacíos por 0, excepto en fechas
    df_libros_seleccionado = df_libros_seleccionado.applymap(lambda x: 0 if pd.isna(x) and not isinstance(x, pd.Timestamp) else x)

    # ✅ Guardar en el diccionario
    datos_extraidos[hoja_libros] = df_libros_seleccionado

    print(df_libros_seleccionado.head())
else:
    print(f"\n⚠️ La hoja '{hoja_libros}' no está en el archivo. Se omite.")

# 📌 Unir todos los datos en un solo DataFrame
df_final = pd.concat(datos_extraidos, names=['Hoja', 'Index'])

# 📌 Guardar en un nuevo archivo Excel
output_file = "datos_filtrados.xlsx"
df_final.to_excel(output_file)

print(f"\n✅ Datos extraídos y guardados en '{output_file}'")
```
